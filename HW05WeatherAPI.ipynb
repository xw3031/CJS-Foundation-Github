**Print only the day with the highest maximum temperature.**

```python
current_max = -9999
max_day = data['timelines']['daily'][0]['time']

for time in data['timelines']['daily']:
    if time['values']['temperatureMax'] > current_max:
        current_max = time['values']['temperatureMax']
        max_day = time['time']
        
print(f'The maximum temperature on {max_day} is {current_max} degrees Celsius.')
```
